# Image to Vector

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
for img in os.listdir('./image'):
    
    img = mpimg.imread('./image/'+img)
    imgplot = plt.imshow(img)
    plt.show()
    print(img.shape)
```

### Add to a vector

```python
X = []

for image in os.listdir('./image'):
    # Ler imagem em Vetor
    X.append(mpimg.imread('./image/'+image))
```

```python
# turn into Array
X = np.array(X)
```
