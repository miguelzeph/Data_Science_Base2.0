```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import tensorflow as tf
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.drop(
    ['sibsp','alone','alive','parch','embarked','deck', 'embark_town','adult_male'],
    inplace=True,
    axis=1
)
df.head(6)
```

```python
df.dropna(inplace=True,axis=0)
```

```python
df.head(6)
```

```python
df['sex'].unique()
```

```python
def sex(x):
    x = x[0]
    if x =='male':
        return 1
    if x =='female':
        return 0
    return x
```

```python
df['sex'] = df[['sex']].apply(sex,axis=1);
```

```python
df['class'].unique()
```

```python
def class_func(x):
    x = x[0]
    if x =='Third':
        return 3
    if x =='Second':
        return 2
    if x =='First':
        return 1
    return x
```

```python
df['class'] = df[['class']].apply(class_func,axis=1);
```

```python
df['who'].unique()
```

```python
def who(x):
    x = x[0]
    if x =='man':
        return 1
    if x =='woman':
        return 0
    if x =='child':
        return -1
    return x
```

```python
df['who'] = df[['who']].apply(who,axis=1);
```

```python
df.head(6)
```

```python
sns.heatmap(df.corr(),annot=True);
```

## Trainamento e Test

```python
x = df.drop('survived',axis=1)
y = df[['survived']]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.3,
)
```

```python
x_train.shape,y_train.shape
```

## SEMPRE RESETAR INDEX DE TUDO! PARA EVITAR PROBLEMAS NO FUTURO

```python
x_train.head()
```

```python
y_train.head()
```

```python
# RESETANDO TODOS OS INDEX
y_train = y_train.reset_index().drop('index',axis=1)
y_train

y_test = y_test.reset_index().drop('index',axis=1)
y_test

x_train = x_train.reset_index().drop('index',axis=1)
x_train

x_test = x_test.reset_index().drop('index',axis=1)
x_test
```

## Modelo

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(
    # Output -> Sai 10
    units=10,
    # Funcão de ativação
    activation = 'relu',
    # Input -> Entra 6 variaveis(Só na primeira)
    input_shape = (x_train.shape[1],)
))

model.add(tf.keras.layers.Dense(
    units =10,
    activation= 'relu',


))

model.add(tf.keras.layers.Dense(
    units=1,
    activation='sigmoid'
))

model.summary()

model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'Adam',
    metrics = ['accuracy']
)
```

```python
history = model.fit(
    x_train,
    y_train,
    validation_split=0.1,
    epochs=100
)

history.history.keys()
```

```python
plt.figure()
plt.plot(history.history['accuracy'],label='acc')
plt.plot(history.history['val_accuracy'],label='acc_val')
plt.legend()
plt.grid()

plt.figure()
plt.plot(history.history['loss'],'.r', label = 'loss')
plt.plot(history.history['val_loss'],'.b', label = 'loss_val')
plt.legend()
plt.grid()
```

## Predictions TREINO

```python
pred_train = model.predict(x_train)
pred_train
```

```python
pred_df = pd.DataFrame()
pred_df['pred'] = pred_train[:,0]
pred_df.head()
```

```python
def cond(x):
    if x>0.5:
        return 1
    return 0

pred_df['pred']=pred_df['pred'].apply(cond)
```

```python
pred_df
```

```python
#pred_df['y'] = y_train
# Problema aqui é que estamos com os índices
# resetados... então ele não vai bater

# Temos que resetar os index de 
# y_train e x_train lá em cima...
#...

# Agora podemos fazer
pred_df['y'] = y_train
```

```python
pred_df
```

```python
k = 0
total = len(pred_df['pred'])

for x,y in zip(pred_df['pred'],pred_df['y']):
    if x == y:
        k = k+1
p = k/total
print(f"Porcentagem de Acerto {np.around(p*100,2)}%")
```
